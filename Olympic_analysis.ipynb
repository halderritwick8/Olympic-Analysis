## 4. Which player has won most number of medals? (Top 5)

```python
import pandas as pd
df = pd.read_csv("summer_olp.csv")
data = []
data1 = []
data2 = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete, len(df[df['Athlete'] == Athlete])]) 
    
data = pd.DataFrame(data, columns = ['Athlete','Medals'])
data = data.sort_values(by = 'Medals', ascending = False).head()
data.plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (7,5),color = 'green')
fun_seg(data,data1,data2)
addlabels(data1,data2)
plt.title("Players won most number of medals")
plt.show()
```
